```python
import numpy as np
def genarate_data():
    return np.random.randn(10000)
```

```python
# use the function genarate_data, to get the 1-d data. like "data = genarate_data()"
# 1. Please normalize (N) and standardize (S) the data and plot (pdf plots) of the same
# 2. Find the expectatoin of both the data N, S
```

```python
data= genarate_data()
print (data)
```

```python
#plot for our random data
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data)
plt.show()
```

```python
#normalising
N=[]
vmin=np.max(data)
vmax=np.min(data)
for i in data:
    N.append(float(i-vmin)/(vmax-vmin))             
sns.kdeplot(np.asarray(N))
plt.show()
 
    
```

```python
#expectation of normalised data
muN=np.mean(N)
print muN
```

```python
#standardisation
S=[]
mu=data.mean()
#print(type(mu))
sigma=np.var(data)**0.5
for i in data:
   S.append(float((i-mu)/sigma))
#print(S)
sns.kdeplot(np.asarray(S))
plt.show()
```

```python
#expectation of standardised data
muN1=np.mean(S)
print muN1
```
